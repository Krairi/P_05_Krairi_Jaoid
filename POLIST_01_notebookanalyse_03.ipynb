### Importer les librairies 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
from zipfile import ZipFile
pd.set_option("display.max_rows", 101)
pd.options.display.max_columns = 999
```

### Importer mon jeu de données créé

```python
z = ZipFile("../P5/data_f.zip")
data = pd.read_csv(z.open('data_f.csv'))

data.shape
```

### Distribution des variables numériques

```python
data.dtypes
```

```python
for column in data.select_dtypes(['int64', 'float64']).columns:
    f, axes = plt.subplots(1,2, figsize=(12,4))
    titre = 'Distribution de ' + str(column)
    plt.title(titre)
    sns.distplot(data[column], bins=30, ax=axes[0])
    titre = 'Distribution de ' + str(column)
    plt.title(titre)
    sns.boxplot(data[column], ax=axes[1])
    plt.show()
```

### Remplacer les valeurs aberrantes

#### Détecter les valeurs aberrante de la variable 'Delai_Moy_Livraison'

```python
sns.boxplot(x=data['Delai_Moy_Livraison'])
```

Le diagramme en moustache détecte un grand nombre de valeur aberrante négative celà ce distingue par des points espacé inférieur à 0.

#### Utilisation de la méthode interquartile  sur la variable 'Delai_Moy_Livraison'

```python
Q1=data['Delai_Moy_Livraison'].quantile(0.25)
Q3=data['Delai_Moy_Livraison'].quantile(0.75)
IQR=Q3-Q1
print(Q1)
print(Q3)
print(IQR)
Lower_Whisker = Q1 - 1.5*IQR
Upper_Whisker = Q3 + 1.5*IQR
print(Lower_Whisker, Upper_Whisker)
```

Cette méthode me permet de relever les points de données qui tombe en dehors de 1,5 fois d’une plage interquartile au-dessus du 3e quartile (Q3= 15) et en dessous du 1er quartile (Q1=6)

#### Traitement des valeurs aberrante 'Delai_Moy_Livraison'

La mèthoque que j'utilise ci-dessous est de définir une variable nommé mask intégrant les valeurs non aberrante, ensuite redéfinir la variable 'Delai_Moy_Livraison' en lui imputant le quartile en pourcentage par tatonement sur les valeurs aberrante.

```python
mask = (data['Delai_Moy_Livraison']>0) & (data['Delai_Moy_Livraison']<Upper_Whisker)
data['Delai_Moy_Livraison'] = data['Delai_Moy_Livraison'].where(mask, data.loc[mask, 'Delai_Moy_Livraison'].quantile(0.5))
```

```python
sns.boxplot(x=data['Delai_Moy_Livraison'])
```
