### Mettre au format les dates

```python
data['order_purchase_timestamp'] = data['order_purchase_timestamp'].astype('datetime64')
data['order_delivered_customer_date'] = data['order_delivered_customer_date'].astype('datetime64')
```

```python
data.info()
```

### Créer la variable delai_dernier_achat

```python
data['Delai_dernier_achat'] = data['order_purchase_timestamp'].max() - data['order_purchase_timestamp']
data['Delai_dernier_achat'] = data['Delai_dernier_achat'].apply(lambda x: x.days)
```

### Créer la variable Delai_Moy_Livraison

```python
data['Delai_Moy_Livraison'] = data['order_delivered_customer_date'] - data['order_purchase_timestamp']
data['Delai_Moy_Livraison'] = data['Delai_Moy_Livraison'].apply(lambda x: x.days)
```

### Créer la variable Heure_achat

```python
data['Heure_achat'] = data['order_purchase_timestamp'].apply(lambda x: x.hour)
```

### Créer la variable Jour_achat

```python
data['Jour_achat'] = data['order_purchase_timestamp'].apply(lambda x: x.weekday())
```

### Créer ma nouvelle base de données

```python
data_f = data.groupby('customer_unique_id').agg(Nb_pdts = pd.NamedAgg(column= 'product_id', aggfunc=np.count_nonzero),
                                                 Nb_com = pd.NamedAgg(column= 'order_id', aggfunc=np.count_nonzero),
                                                 Mont_max_achats = pd.NamedAgg(column= 'price', aggfunc=sum),
                                                 Mont_moy_achats = pd.NamedAgg(column= 'price', aggfunc=np.mean), 
                                                 Note_Moy_Com = pd.NamedAgg(column= 'review_score', aggfunc=np.mean),
                                                Date_dernier_achat = pd.NamedAgg(column= 'order_purchase_timestamp', aggfunc=max),
                                                 Villes = pd.NamedAgg(column= 'customer_city', aggfunc=np.unique),
                                                 Cat_prod = pd.NamedAgg(column= 'product_category_name_english',aggfunc=np.unique),
                                                Delai_moy_dernier_achat = pd.NamedAgg(column= 'Delai_dernier_achat',aggfunc=np.mean),
                                               Delai_Moy_Livraison = pd.NamedAgg(column= 'Delai_Moy_Livraison',aggfunc=np.mean),
                                               Heure_achat_moy = pd.NamedAgg(column= 'Heure_achat',aggfunc=np.mean),
                                               Jour_achat_moy = pd.NamedAgg(column= 'Jour_achat',aggfunc=np.mean),
                                               Type_paiement = pd.NamedAgg(column= 'payment_type',aggfunc=np.unique),
                                               Note_facil_pai_moy = pd.NamedAgg(column= 'payment_installments',aggfunc=np.mean),
                                               Latitude = pd.NamedAgg(column= 'geolocation_lat',aggfunc=np.mean),
                                               Longitude = pd.NamedAgg(column= 'geolocation_lng',aggfunc=np.mean))
```
